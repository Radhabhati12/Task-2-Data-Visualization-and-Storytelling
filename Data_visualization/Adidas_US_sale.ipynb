```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_excel("C:\\Users\\radhe\\OneDrive\\Desktop\\Adidas US Sales Datasets.xlsx", engine='openpyxl')
df
```

```python
df['Total Sales'].count()
```

```python
Profit_percentage= (df['Operating Profit'].sum()/df['Total Sales'].sum())*100
print('Profit Percentage =',Profit_percentage)
```

```python
pd.crosstab(df['Retailer'], df['Sales Method'])
```

```python
sns.barplot(x='Region', y='Price per Unit', data=df, color= 'cyan', hue='Sales Method')
plt.title('Price per unit in different region', fontweight='bold')
plt.show()
```

```python
sns.barplot(x='Retailer', y='Operating Profit',data=df, palette= 'winter')
plt.title('Profit by region',fontweight='bold')
plt.show()
```

```python
sns.countplot(x='Retailer', data=df, hue='Sales Method')
plt.title(' no of sales methods for Retailer', fontweight='bold')
plt.show()
```

```python
sns.countplot(x='Region', data=df, hue='Sales Method', palette='twilight')
plt.title(' No of sales methods in different Region', fontweight='bold')
plt.show()
```
